```python
import matplotlib.pyplot as plt
import numpy as np
import math as m
```

```python
plt.figure(figsize = (10,10))


plt.title('Big O Nation')
plt.xlabel('Elements')
plt.ylabel('Operations')

x = range(1,20)

y1 = [(0* x +1) for x in x]
y2 = [m.log(x) for x in x]
y3 = [x for x in x]
y4 = [(x * m.log(x)) for x in x]
y5 = [x**2 for x in x]
y6 = [2**x for x in x]
y7 = [m.factorial(x) for x in x]
y8 = [(x**x) for x in x]


plt.plot(x, y1, c = 'gray', label = 'O(1)$', ls = '--', marker = 'v')
plt.plot(x, y2, c = 'black', label='O(log(n)', marker = '^')
plt.plot(x, y3, c = 'blue', label='O(n)', marker = 'o')
plt.plot(x, y4, c = 'green', label='$O(n log n)$', )
plt.plot(x, y5, c = 'gray', label = '$O(n^2)$', ls = '--', marker = 'v')
plt.plot(x, y6, c = 'brown', label='$O(2^n)$', marker = '^')
plt.plot(x, y7, c = 'magenta', label='$O(n!)$', marker = 'o')
plt.plot(x, y8, c = 'orange', label='$O(n^n)$')

plt.ylim(0, 20)
plt.xlim(0, 20)

plt.legend()
```
