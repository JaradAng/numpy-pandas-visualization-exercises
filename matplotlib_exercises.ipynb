```python
import matplotlib.pyplot as plt
import numpy as np

#y=x^2−x+2
```

```python
x_axis = range(-50, 50)
y_axis = [n**2 -n+2 for n in x_axis]

plt.figure(figsize = (10, 10))

plt.title('$y=x^2−x+2$')


plt.plot(x_axis, y_axis)

plt.annotate('origin', xy = (0,0), arrowprops = {})
```

```python
x = range(1, 50)
y = [x**0.5 for x in x]



plt.figure(figsize=(10,10))
plt.title('y = √x ')
plt.plot(x, y)


plt.xlabel('x')
plt.ylabel('√x')
```

```python
x = range(-100, 100)
y = [x**3 for x in x]

plt.figure(figsize= (10, 10))

plt.title('$y = x^3$')

plt.plot(x,y)

plt.xlabel('x')
plt.ylabel('$x^3$')
```

```python
x = range(-10, 100)
y = [2**x for x in x]

plt.figure(figsize= (10, 10))

plt.title('$ y = 2^x$')

plt.plot(x,y)

plt.xlabel('x')
plt.ylabel('$2^x$')
```

```python
x = range(0, 100)
y = [(1/ (x +1)) for x in x]

plt.figure(figsize= (10, 10))

plt.title('$y = 1/(x+1)$')

plt.plot(x,y)

plt.xlabel('x')
plt.ylabel('$1/(x+1)$')
```

```python
plt.figure(figsize = (10,15))
plt.suptitle('all as subplot')

##Subplot 1
plt.subplot(411)
x = range(-100, 1000)
y = [x**0.5 for x in x]
plt.title('$y = √x $')

plt.plot(x, y)

##subplot 2
plt.subplot(412)
x = range(-1000, 1000)
y = [x**3 for x in x]
plt.plot(x,y)
plt.title('$y = x^3$')


##subplot3
plt.subplot(413)
x = range(-10, 100)
y = [2**x for x in x]
plt.title('$ y = 2^x$')
plt.plot(x, y)

##subplot 4
plt.subplot(414)
x = range(0, 100)
y = [(1/ (x +1)) for x in x]
plt.plot(x, y)

plt.title('$y = 1/(x+1)$')


## played around with the sizing and how to lay out the subplots. ended up with the enlonged graphs
```

```python
fig = plt.figure(figsize = (10,10))


x = list(range(0,20))

# define multiple y's
y1 = [x**0.5 for x in x]
y2 = [x**3 for x in x]
y3 = [2**x for x in x]
y4 = [(1/ (x +1)) for x in x]

plt.plot(x, y1, c = 'gray', label = '$\sqrt{x}$', ls = '--', marker = 'v')
plt.plot(x, y2, c = 'black', label='$x^{3}$', marker = '^')
plt.plot(x, y3, c = 'blue', label='$2^{x}$', marker = 'o')
plt.plot(x, y4, c = 'green', label='$1/(x+1)$' )

plt.ylim(0, 10)
plt.xlim(0, 5)

plt.title('4 Different function')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()


```
